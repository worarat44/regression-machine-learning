การประเมินโมเดล Regression มี 4 แบบ ดังนี้ 
1) Mean Absolute Error (MAE) การหาผลรวมระยะคลาดเคลื่อน คือหาว่าโมเดลมีความแม่นยำในการพยากรณ์มากน้อยเพียงใด ถ้า MAE มีค่าน้อยแสดงว่าคลาดเคลื่อนน้อย หมายความว่าโมเดลมีความแม่นยำสูง แต่ถ้าค่า MAE มาก แสดงว่าโมเดลมีความคลาดเคลื่อนมาก คือโมเดลมีความแม่นยำน้อย 
2) Mean Squared Error (MSE) การหาผลรวมระยะคลาดเคลื่อนด้วยวิธียกกำลังสอง การแปลผลเหมือนกับข้อ 1
3) Root Mean Squared Error (RMSE) การหาผลรวมระยะคลาดเคลื่อนด้วยวิธีการถอดรูทรากที่สอง การแปลผลเหมือนกับข้อ 1
4) Coefficients of Determination การหาค่าสัมประสิทธิ์การตัดสินใจ หรือ R square (R2) มีค่าระหว่าง 0 - 1 ถ้าค่าเข้าใกล้ 1 หรือเท่ากับ 1 แสดงว่าโมเดลมีความแม่นยำสูง แต่ถ้าค่าเข้าใกล้ 0 แสดงว่าโมเดลมีความแม่นยำต่ำ

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print('ผลรวมระยะคลาดเคลื่อนเท่ากับ', mean_absolute_error(y_test, y_predict).round(2))
print('ผลรวมระยะคลาดเคลื่อนด้วยวิธียกกำลังสองเท่ากับ', mean_squared_error(y_test, y_predict).round(2))
print('ผลรวมระยะคลาดเคลื่อนด้วยวิธีการถอดรูทรากที่สองเท่ากับ', np.sqrt(mean_squared_error(y_test, y_predict)).round(2))
print('การหาค่าสัมประสิทธิ์การตัดสินใจ หรือ R square (R2) เท่ากับ', model.score(X_train, y_train).round(2))
#print('การหาค่าสัมประสิทธิ์การตัดสินใจ หรือ R square (R2) เท่ากับ', model.score(X, y).round(2))
```

การแปลผล
1) ถ้าค่า p (P>|t|) < 0.05 แสดงว่ามีความสัมพันธ์กัน คือคือตัวแปรต้น (X หรือเหตุ) เป็นสาเหตุ/ส่งผลต่อตัวแปรตาม (Y หรือผล) เช่น กินจุ (เหตุ) ทำให้อ้วน (ผล) ขณะเดียวกัน อ้วน (เหตุ) ทำให้เป็นโรคความดัน เบาหวาน (ผล) เป็นต้น
2) ถ้าค่า p (P>|t|) > 0.05 แสดงว่าไม่มีความสัมพันธ์กัน คือคือตัวแปรต้น (X หรือเหตุ) ไม่เป็นสาเหตุ/ไม่ส่งผลต่อตัวแปรตาม (Y หรือผล) เช่น กินจุ (เหตุ) แต่ก็ไม่ทำให้อ้วน (ผล) ขณะเดียวกัน อ้วน (เหตุ) ไม่ทำให้เป็นโรคความดัน เบาหวาน (ผล) เป็นต้น
3) ค่าอย่างมีนัยสำคัญทางสถิติที่..... ถ้าค่า p < 0.05 แต่ > 0.01 อย่างมีนัยสำคัญทางสถิติที่ 0.05, ถ้าค่า p < 0.01 แต่ > 0.001 อย่างมีนัยสำคัญทางสถิติที่ 0.01, ถ้าค่า p < 0.001 อย่างมีนัยสำคัญทางสถิติที่ 0.000

### แบบจำลองในการศึกษา 
### Y = a + b1x1 + b2x2 + b3x3 + b4x4 + b5x5 + e
### กำหนดให้
#### a = ค่าคงที่ความถดถอยพหุ
#### b1 ... b5 = ค่าสัมประสิทธิถดถอยของตัวแปรอิสระ (ตัวแปรต้น)
#### Y = รายจ่ายสาธารณะ (% GDP)
#### x1 = จีดีพีต่อหัว
#### x2 = สัดส่วนประชากรที่มีความยากจน
#### x3 = ภาษีสินค้าและบริการ
#### x4 = ดัชนีคอรัปชัน
#### x5 = ภาษีรายได้
#### e = ค่าความคลาดเคลื่อน

## การอธิบายแปลผล เอาเฉพาะตัวที่มีนัยสำคัญทางสถิติ คือค่า p < 0.05 เท่านั้นมาอธิบายแปลผล

### รายจ่ายสาธารณะ = 4.8051 + -0.0003(x1) + -0.1191(x2) + -0.0673(x3) + 0.1542(x4) + 0.9430(x5) + 4.974 คือพยากรณ์ผิดพลาดจากข้อมูลจริงคลาดเคลื่อนออกไป 4.974

### ผลการศึกษาพบว่า จีดีพีต่อหัว (x1) มีความสัมพันธ์ในทิศทางลบกับ รายจ่ายสาธารณะ (Y) ที่ระดับนัยสำคัญทางสถิติ 0.000 โดยมีค่าสัมประสิทธิ์จีดีพีต่อหัว เท่ากับ -0.0003 หมายความว่า ถ้าจีดีพีต่อหัว เพิ่มขึ้น 1 หน่วย ส่งผลทำให้รายจ่ายสาธารณะ ลดลง 0.0003 หน่วย 

### ผลการศึกษาพบว่า  ดัชนีคอรัปชัน (x4) มีความสัมพันธ์ในทิศทางบวกกับ รายจ่ายสาธารณะ (Y) ที่ระดับนัยสำคัญทางสถิติ 0.004 โดยมีค่าสัมประสิทธิ์ดัชนีคอรัปชัน เท่ากับ 0.1542 หมายความว่า ดัชนีคอรัปชัน เพิ่มขึ้น 1 หน่วย ส่งผลทำให้รายจ่ายสาธารณะ เพิ่มขึ้น 0.1542 หน่วย 


### ผลการศึกษาพบว่า ภาษีรายได้ (x5) มีความสัมพันธ์ในทิศทางบวกกับ รายจ่ายสาธารณะ (Y) ที่ระดับนัยสำคัญทางสถิติ 0.000 โดยมีค่าสัมประสิทธิ์ภาษีรายได้ เท่ากับ 0.9430 หมายความว่า ภาษีรายได้ เพิ่มขึ้น 1 หน่วย ส่งผลทำให้รายจ่ายสาธารณะ เพิ่มขึ้น 0.9430 หน่วย
